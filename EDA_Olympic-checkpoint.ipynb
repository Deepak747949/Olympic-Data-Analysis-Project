# Olympic 2024

**Objective :**


To analyze the Olympic dataset to understand trends in participation and performance, explore how medals are distributed among countries, sports, and athletes, examine gender representation and its changes over the years, and identify how age impacts athletic performance.

```python
#Importing the required library
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
```

```python
#Loading the dataset from file
athletes=pd.read_csv('Dataset/Source/athletes.csv')
events=pd.read_csv('Dataset/Source/events.csv')
medallists=pd.read_csv('Dataset/Source/medals.csv')
medals_total=pd.read_csv('Dataset/Source/medals_total.csv')
history_data=pd.read_csv('Dataset/Historical/Olympic_Games_Medal_Tally.csv')
Oly=pd.read_csv('Dataset/Historical/athlete_events.csv')
```

Working With Athelets Data

```python
#Getting the firtst five rows of data
athletes.head()
```

```python
#Getting the overvieww of data
athletes.info()
```

```python
#Getting the copy of important column form athletes data
athlete=pd.DataFrame(athletes[['name','gender','country','disciplines','events','birth_date','birth_country']])
```

```python
#Getting the first five row form data
athlete.head()
```

```python
#Checking for the duplicate value
athlete.duplicated().sum()
```

Missing value is not present in our data

```python
#Checking the Missing value in data
(athlete.isna().sum()/athlete.shape[0])*100
```

There is 14.84% Missing value present in birth_country Column in our data

```python
#Filling the Missing value with Unknown in our data
athlete['birth_country']=athlete['birth_country'].fillna('Unknown')
```

```python
#Cleaning the disciplines column in our data
athlete['disciplines']=athlete['disciplines'].str.replace('[','')
athlete['disciplines']=athlete['disciplines'].str.replace(']','')
athlete['disciplines']=athlete['disciplines'].str.replace('"','')
athlete['disciplines']=athlete['disciplines'].str.replace("'",'')
```

```python
#Creating a new column Age from date of birth in our data
athlete['age'] = (pd.to_datetime('today') - pd.to_datetime(athlete['birth_date'], dayfirst=True)).dt.days // 365
```

```python
#Analyzing the Age of the athelets
age=athlete['age'].value_counts()
plt.figure(figsize=(20,8))
plt.hist(athlete['age'],edgecolor='black', alpha=0.8, color='skyblue')
plt.title('Age Distribution of Athletes')
plt.xlabel('Age')
plt.ylabel('Number of Athletes') 
plt.show()
```

**Insights**

Most of the Athelete is in between 18-35 Age

```python
#Analyzing the Gender of athelete
gender=athlete['gender'].value_counts()
plt.figure(figsize=(20,8))
fig=plt.barh(gender.index,gender.values,color=['lightpink','lightblue'])
plt.title("The Gender and There count")
plt.xlabel('Total Number')
plt.ylabel("Gender")
plt.bar_label(fig,labels=gender.values,padding=2)
plt.show()
```

**Insights**

The graph shows higher participation of male athletes compared to female athletes

```python
#Analyze the Country with number of athlete
country=athlete.country.value_counts().head(20)
warnings.filterwarnings('ignore')
plt.figure(figsize=(20,8))
sns.barplot(x=country.index,y=country.values,palette='plasma')
plt.xticks(rotation=20)
plt.title("The Top 20 Country By Athlete",fontsize=22)
plt.xlabel("Country")
plt.ylabel("The Number of Athlete")
plt.show()
warnings.filterwarnings('ignore')
```

```python
#Analyzing the medal distribution by countyr and medal type
medals_total = medals_total.head(20)
# Reshape the data to a long format using pd.melt
medals_long = pd.melt(medals_total, 
                      id_vars=["country_code"],  # Columns to keep as identifiers
                      value_vars=["Gold Medal", "Silver Medal", "Bronze Medal"],  # Columns to unpivot
                      var_name="Medal Type",  # Name for the column that will hold medal types
                      value_name="Count") 

# Create the bar plot with Seaborn
plt.figure(figsize=(20,8))
sns.barplot(x="country_code", y="Count",
            hue="Medal Type", data=medals_long,
            palette={"Gold Medal": "#FFD700", "Silver Medal": "#C0C0C0", "Bronze Medal": "#CD7F32"}
           )

# Rotate x-axis labels
plt.xticks(rotation=0)

# Add title and labels
plt.title("Top 20 County by Medal Counts", fontsize=16)
plt.xlabel("Country Code", fontsize=14)
plt.ylabel("Medal Count", fontsize=14)
plt.show()
```

**Insights**
* USA and China have Highest Gold Medal
* USA China and France have highest Silver Medal
* USA and GBR(United Kingdom) have higheset Bronze Medal


```python
# Grouping and aggregating the data
df = medallists.groupby(['country_code', 'gender'])['medal_type'].count().reset_index()

# Sorting by total medal count (descending order)
df['total_medals'] = df.groupby('country_code')['medal_type'].transform('sum')
df = df.sort_values(by='total_medals', ascending=False).drop(columns='total_medals')

# Filter for top countries
top_countries = df['country_code'].value_counts().head(10).index
df = df[df['country_code'].isin(top_countries)]
custom_palette = {
    "M": "blue",
    "W": "pink",
    "O": "grey",
    "X":'green'# Example for an additional gender
}

# Plot using Seaborn
plt.figure(figsize=(20,8))
sns.barplot(
    x="country_code", 
    y="medal_type", 
    hue="gender", 
    data=df, 
    palette=custom_palette
)

# Adding titles and labels
plt.title("Country-wise and Gender-wise Medal Distribution", fontsize=16)
plt.xlabel("Country", fontsize=14)
plt.ylabel("Medal Count", fontsize=14)
plt.xticks(rotation=0)
plt.legend(title="Gender", fontsize=12)

# Display the plot
plt.tight_layout()
plt.show()
```

**Insights**
* Most of the Female from USA China Austrila and GBR(United Kingdom) have won the medal
* Most of the Male from USA, France, China and GBR(United Kingdom) have won the medal

**Analyzing India at Olympic**

```python
#Filtering and Analyzing india participated at Olympic gender-wise


year_wise=ind_ath.groupby('Year')['Sex'].value_counts()
year_wise_df = year_wise.reset_index(name='Count')

#Adding the data of olympic 2020,2022 and 2024
new_rows = [
    (2020, 'M', 67),
    (2020, 'F', 52),
    (2022, 'M', 1),
    (2024, 'M', 70),
    (2024, 'F', 47)]

new_data = pd.DataFrame(new_rows, columns=['Year', 'Sex', 'Count'])
# Convert existing Series to DataFrame for easier merging
year_wise_df = year_wise.reset_index(name='Count')
year_wise_df.columns = ['Year', 'Sex', 'Count']

# Append new rows and reindex
year_wise_df = pd.concat([year_wise_df, new_data], ignore_index=True)
year_wise_df = year_wise_df.sort_values(by=['Year', 'Sex']).reset_index(drop=True)

# Convert back to Series with MultiIndex
year_wise = year_wise_df.set_index(['Year', 'Sex'])['Count']
```

```python
# Plot using seaborn
plt.figure(figsize=(20,7))
sns.barplot(data=year_wise_df, x="Year", y="Count", hue="Sex", palette={"M": "blue", "F": "pink"})

# Customize the plot
plt.title("Year-wise Gender Participation of India", fontsize=16)
plt.xlabel("Year", fontsize=14)
plt.ylabel("Number of Participants", fontsize=14)
plt.xticks(rotation=45)
plt.legend(title="Gender", title_fontsize=12, fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Show the plot
plt.tight_layout()
plt.show()
```

**Insights**
* Male Participants are rising since last Decade.
* Female Participants was peak in 2016 

```python
#Analyzing the Medal Of all time of India 
ind_his=history_data.loc[history_data['country']=='India']

medals_long = pd.melt(ind_his, 
                      id_vars=["year", "country_noc"],  # Columns to keep as identifiers
                      value_vars=["gold", "silver", "bronze"],  # Columns to unpivot
                      var_name="Medal Type",  # Name for the column that will hold medal types
                      value_name="Count") 

# Rename the medal types to more descriptive names
medals_long["Medal Type"] = medals_long["Medal Type"].str.capitalize() + " Medal"

# Create the bar plot with Seaborn
plt.figure(figsize=(20,8))
sns.barplot(x="year", y="Count",
            hue="Medal Type", data=medals_long,
            palette={"Gold Medal": "#FFD700", "Silver Medal": "#C0C0C0", "Bronze Medal": "#CD7F32"}
           )

# Rotate x-axis labels
plt.xticks(rotation=0)

# Add title and labels
plt.title("India's Olympic Medal Counts by Year", fontsize=16)
plt.xlabel("Year", fontsize=14)
plt.ylabel("Medal Count", fontsize=14)
plt.show()
```

**Insights**
* India's Olympic performance peaked in the 2020 Tokyo Olympics, winning a record 7 medals
